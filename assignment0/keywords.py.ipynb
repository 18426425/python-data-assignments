```python
import os
txt = os.listdir(r'///Users/amberyu/Desktop/assignment0/')
print (txt)
```

```python
all_content = ''

for info in txt:
    domain = os.path.abspath(r'///Users/amberyu/Desktop/assignment0/')
    info = os.path.join(domain, info)
    fr = open (info, 'r+')
    content = fr.read()
    all_content += content
```

```python
words = str.split(all_content)
```

```python
type (words)
```

```python
stop_words = ['the', 'and', 'of', 'a', 'in','to','as','is','on','that','with','The','for','it','has','are','not','have','its','from','would','be','at','an','which','by','this','up','said','also','will','more','about','was','last','been','they','see','if','In','I','but','he']
for i in words:
    if i in stop_words:
        words.remove(i)
```

```python
words
```

```python
len(words)
```

```python
count = []
for x in words:
    times = words.count(x)
    count.append(times)
```

```python
words_count = dict(zip(words, count))
```

```python
words_count
```

```python
rank = sorted(words_count.items(), key=lambda x:x[1],  reverse=True)  
```

```python
rank
```

```python
top_15 = rank[:15]
```

```python
top_15
```

```python
import csv
with open('keywordsfrequency.csv', 'w') as csvfile:
    writer = csv.writer(csvfile)
    header = ['keyword','frequency']
    writer.writerow(header)
    writer.writerows(top_15)
```
